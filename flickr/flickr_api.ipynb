```python
## 1. 필요한 모듈 임포트
import flickrapi
import urllib.request
import time
import os

## 2. api키 및 비밀번호를 셋팅하고 flickr 객체를 생성한다.
# - [flickr api키 발급페이지](https://www.flickr.com/services/apps/create/)
api_key = u'fc9b1b09eaf69edb5347c866c1ed9957'
api_secret = u'36655c3bf56474e2'
flickr = flickrapi.FlickrAPI(api_key, api_secret, format='parsed-json')

## 3. 각종 파라미터를 설정한다.
TEXT = 'food'
PAGE = 1
SAVE_FOLDER_NAME = 'food'
```

```python
## 4. flickr.photos.search api를 이용하여 이미지의 url를 추출한다.
before_items = ""
img_urls = []

while True:
    photos = flickr.photos.search(text=TEXT,        # 검색어 설정
                                  page=PAGE,        # 조회할 페이지 번호 설정
                                  sort="relevance", # 관련도순 정렬
                                  extras="url_o")   # 이미지의 url을 바로 리턴해주므로 필수 파라미터임
    
    items = photos['photos']['photo']

    if before_items == items:
        print("*** 중복 정보가 수집되어서 id 스크래핑을 종료합니다. ***\n")
        break

    temp_img_url = [item['url_o'] for item in items if 'url_o' in item.keys()]
    img_urls.extend(temp_img_url)
    print("이미지 url 수집:  {}번째 페이지까지 완료되었습니다.  {}".format(PAGE, temp_img_url[0]))
    
    before_items = items
    PAGE += 1
    
    time.sleep(3)
    
img_urls = list(set(img_urls))
print("수집된 이미지의 개수: ", len(img_urls))
```

```python
## 5. 추출된 이미지 url을 이용하여 이미지를 원하는 경로에 저장한다.
# urllib.request.urlretrieve(다운로드 할 이미지 URL, 저장할 경로 및 파일명)

if SAVE_FOLDER_NAME not in os.listdir():
    os.mkdir(SAVE_FOLDER_NAME)

for img_url in img_urls:
    urllib.request.urlretrieve(img_url, './{}/{}_{}.jpg'.format(SAVE_FOLDER_NAME, TEXT, img_urls.index(img_url)+1))
    print("{}(/{})번째 이미지를 지정 폴더에 저장하였습니다.".format(img_urls.index(img_url) + 1, len(img_urls)))
```
